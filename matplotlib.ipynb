Scipy:
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months = np.arange(12)
```

```python
plt.scatter(months, max_temp)
plt.scatter(months, min_temp)

plt.xlabel("Month")
plt.ylabel("Temp(*C)")
plt.legend
plt.show()
```

This assignment is for visualization using matplotlib:

Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic
```

```python
titanic = titanic.drop(index = 1309)
```

```python
plt.pie(titanic["sex"].value_counts(), autopct = '%0.2f%%', labels = ["male", "female"])
plt.show()
```

```python
colors = {"male":"blue", "female":"red"}
plt.scatter(titanic["fare"], titanic["age"], c = titanic["sex"].dropna().map(colors))

plt.xlabel("fare")
plt.ylabel("age")
plt.show()
```
